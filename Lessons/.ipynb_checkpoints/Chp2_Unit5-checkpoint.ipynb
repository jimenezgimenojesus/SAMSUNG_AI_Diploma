### Ajuste (no lineal) por mínimos cuadrados: método del gradiente descendiente.


Cuando la nube de puntos que queremos aproximar no es lineal, o el número de datos es considerablemente grande, el ajuste lineal por el método de mínimos cuadrados no resulta efectivo...

Veamos un método alternativo: si nos fijamos, lo que tenemos es una nube de puntos $(x_i, y_i)$ con $i\geq 0$ posiblemente grande. En el caso de que quisiéramos ajustar esta nube de puntos por una función $y=f(x)$, entonces lo que queremos es minimizar el error cometido entre $y_i$ y nuestra predicción $\hat{y_i}=f(x_i)$. Hay varias maneras de medir el error cometido, como el _error cuadrático medio_, pero usaremos aquí simplemente el error cuadrático:

$$\sum_i (y_i-\hat{y_i})^2 = \sum_i (y_i-f(x_i))^2.$$

Luego tenemos que hallar el **mínimo de una función de varias variables**.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as la
```

```python
xi = np.array([-2, -1, 0, 1, 2, 3, 4, 5])
yi = np.array([-2, -1, 1.5, -2, 2, 1, 3.5, 3])

plt.figure(figsize=(4, 4))
plt.xlim(-2.5, 5.5)
plt.ylim(-4, 14)
plt.plot(xi, yi, '*g')
plt.show()
```

```python
# probemos un ajuste lineal del tipo y = a*x + b, por ejemplo y = 2x+2

a, b = 2, 2
y_gorrito = a*xi + b

t = np.linspace(-2.5, 5.5, 1000)

plt.figure(figsize=(4,4))
plt.xlim(-2.5, 5.5)
plt.plot(xi, yi, '*g', xi, y_gorrito, 'or', t, 2*t+2, '-r')
plt.show()
```

```python
# Error cometido

error = np.sum((yi - y_gorrito)**2)
print('Éste es el error cuadrático:', error)
```

```python
# Ajuste lineal

a, b = 0.723, -0.355

y_gorrito = a * xi + b


plt.figure(figsize=(4, 4))
plt.xlim(-2.5, 5.5)
plt.ylim(-4, 6)
plt.plot(xi, yi, '*g', xi, y_gorrito, 'or', t, a*t+b, 'r')
plt.show()

# Error cometido

error = np.sum((yi - y_gorrito)**2)
print('Éste es el error cuadrático:', error)
```

```python
# Ajuste exponencial del tipo y = a*e^(bx)

a, b, c = 0.56, 0.45, -1

y_gorrito = a * np.exp(b * xi) + c


plt.figure(figsize=(4, 4))
plt.xlim(-2.5, 5.5)
plt.ylim(-2.5, 5)
plt.plot(xi, yi, '*g', xi, y_gorrito, 'or', t, a*np.exp(b*t)+c, 'r')
plt.show()

# Error cometido

error = np.sum((yi - y_gorrito)**2)
print('Éste es el error cuadrático (INICIAL):', error)
```

```python
# Método del Gradiente Descendiente (ajuste lineal)

xbarra = np.array([[2], [2]])
alpha = 0.005
gradiente = np.array([[np.sum(2 * (yi - (xbarra[0, 0]*xi + xbarra[1,0]))*(-xi))],
                       [np.sum(2 * (yi - (xbarra[0, 0]*xi + xbarra[1,0])) * (-1))]])

# Tener cuidado con CÓMO SE PONEN los valores, si por columna o fila [[]]
xbarra = xbarra - alpha * gradiente

# Error cometido
error = np.sum((yi - (xbarra[0, 0]*xi + xbarra[1,0]))**2)
# El error tiene que ser MÁS PEQUEÑO después de la primera iteración
print('Éste es el error cuadrático (INICIAL):', error)
print(xbarra, end='\n\n')

for i in range (100):
    gradiente = np.array([[np.sum(2 * (yi - (xbarra[0, 0]*xi + xbarra[1,0]))*(-xi))],
                          [np.sum(2 * (yi - (xbarra[0, 0]*xi + xbarra[1,0])) * (-1))]])
    xbarra = xbarra - alpha * gradiente
    
    # Error cometido
    error = np.sum((yi - (xbarra[0, 0]*xi + xbarra[1,0]))**2)
    # El error tiene que ser MÁS PEQUEÑO después de la primera iteración
    # print('Éste es el error cuadrático:', error)

print('Éste es el error cuadrático (FINAL):', error)
print(); print(xbarra)
```

```python
# Ajuste lineal

a, b = 0.71265568, -0.31400706
y_gorrito = a * xi + b

plt.figure(figsize=(4, 4))
plt.xlim(-2.5, 5.5)
plt.ylim(-4, 6)
plt.plot(xi, yi, '*g', xi, y_gorrito, 'or', t, a*t+b, 'r')
plt.show()

# Error cometido
error = np.sum((yi - y_gorrito)**2)
print('Éste es el error cuadrático:', error)

# Si quieres utilizar un ALPHA grande (como 0.1), normaliza el vector gradiente (divídelo entre la raíz de la suma de los cuadrados de sus componentes)
```

```python
# Método del Gradiente Descendiente (ajuste lineal)

xbarra = np.array([[0.5], [0.4], [-1]])
alpha = 0.01
gradiente = np.array([[np.sum(2 * (yi - (xbarra[0, 0] * np.exp(xbarra[1, 0]*xi)+xbarra[2, 0]))*(-np.exp(xbarra[1,0]*xi)))],
                      [np.sum(2 * (yi - (xbarra[0, 0] * np.exp(xbarra[1, 0]*xi)+xbarra[2, 0]))*(-xbarra[0,0]*np.exp(xbarra[1, 0]*xi)*xi))],
                      [np.sum(2 * (yi - (xbarra[0, 0] * np.exp(xbarra[1, 0]*xi)+xbarra[2, 0]))*(-1))]])

# Tener cuidado con CÓMO SE PONEN los valores, si por columna o fila [[]]
xbarra = xbarra - alpha * gradiente/la.norm(gradiente)

# Error cometido
error = np.sum((yi - (xbarra[0, 0]*np.exp(xbarra[1,0]*xi)+xbarra[2, 0]))**2)
# El error tiene que ser MÁS PEQUEÑO después de la primera iteración
print('Éste es el error cuadrático:', error)
print(xbarra, end='\n\n')

for i in range (10000):
    gradiente = np.array([[np.sum(2 * (yi - (xbarra[0, 0] * np.exp(xbarra[1, 0]*xi)+xbarra[2, 0]))*(-np.exp(xbarra[1,0]*xi)))],
                          [np.sum(2 * (yi - (xbarra[0, 0] * np.exp(xbarra[1, 0]*xi)+xbarra[2, 0]))*(-xbarra[0,0]*np.exp(xbarra[1, 0]*xi)*xi))],
                          [np.sum(2 * (yi - (xbarra[0, 0] * np.exp(xbarra[1, 0]*xi)+xbarra[2, 0]))*(-1))]])
    
    xbarra = xbarra - alpha * gradiente/la.norm(gradiente)
    
    # Error cometido
    error = np.sum((yi - (xbarra[0, 0]*np.exp(xbarra[1,0]*xi)+xbarra[2, 0]))**2)
    # El error tiene que ser MÁS PEQUEÑO después de la primera iteración
    # print('Éste es el error cuadrático:', error)

print()
print('Éste es el error cuadrático:', error)
print(xbarra)
```

```python
# Ajuste exponencial del tipo y = a*e^(bx)

a, b, c = 6.2902933, 0.0907844, -6.68865564
y_gorrito = a * np.exp(b * xi) + c

plt.figure(figsize=(4, 4))
plt.xlim(-2.5, 5.5)
plt.ylim(-2.5, 5)
plt.plot(xi, yi, '*g', xi, y_gorrito, 'or', t, a*np.exp(b*t)+c, 'r')
plt.show()

# Error cometido
error = np.sum((yi - (xbarra[0, 0]*np.exp(xbarra[1,0]*xi)+xbarra[2, 0]))**2)
print('Éste es el error cuadrático (INICIAL):', error)
```

```python
# Método del Gradiente Descendiente (ajuste exponencial)

xi = np.array([0.0, 1.0, 1.5, 2.0, 2.5, 3.0, 6, 6.5])
yi = np.array([0.10, 0.85, 0.99, 0.91, 0.60, 0.12, -0.05, 0.42])

xbarra = np.array([[4.0], [1.0], [1]])
alpha = 0.001
gradiente = np.array([[np.sum(2 * (yi -(xbarra[0, 0]*np.sin(np.exp(xbarra[1,0]*xi))+xbarra[2, 0])) * (- np.sin(xbarra[1,0]*xi)))],
                      [np.sum(2 * (yi -(xbarra[0, 0]*np.sin(np.exp(xbarra[1,0]*xi))+xbarra[2, 0])) * (- xbarra[0,0]*xi*np.cos(xbarra[1,0]*xi)))],
                      [np.sum(2 * (yi -(xbarra[0, 0]*np.sin(np.exp(xbarra[1,0]*xi))+xbarra[2, 0])) * (- 1))]])

# Tener cuidado con CÓMO SE PONEN los valores, si por columna o fila [[]]
xbarra = xbarra - alpha * gradiente/la.norm(gradiente)

# Error cometido
error = np.sum((yi - (xbarra[0, 0]*np.sin(np.exp(xbarra[1,0]*xi))+xbarra[2, 0]))**2)
# El error tiene que ser MÁS PEQUEÑO después de la primera iteración
print('Éste es el error cuadrático (INICIAL):', error)
print(xbarra, end='\n\n')

for i in range (1000):
    gradiente = np.array([[np.sum(2 * (yi -(xbarra[0, 0]*np.sin(np.exp(xbarra[1,0]*xi))+xbarra[2, 0])) * (- np.sin(xbarra[1,0]*xi)))],
                          [np.sum(2 * (yi -(xbarra[0, 0]*np.sin(np.exp(xbarra[1,0]*xi))+xbarra[2, 0])) * (- xbarra[0,0]*xi*np.cos(xbarra[1,0]*xi)))],
                          [np.sum(2 * (yi -(xbarra[0, 0]*np.sin(np.exp(xbarra[1,0]*xi))+xbarra[2, 0])) * (- 1))]])
    
    xbarra = xbarra - alpha * gradiente/la.norm(gradiente)
    
    # Error cometido
    error = np.sum((yi - (xbarra[0, 0]*np.sin(np.exp(xbarra[1,0]*xi))+xbarra[2, 0]))**2)
    # El error tiene que ser MÁS PEQUEÑO después de la primera iteración
    # print('Éste es el error cuadrático:', error)

print()
print('Éste es el error cuadrático (FINAL):', error)
print(xbarra)
```

```python
# Ajuste sinusoidal del tipo y = a*e^(bx)

a, b, c = 4.0, 1, 1
y_gorrito = xbarra[0,0]*np.sin(xbarra[2,0]*xi)+c

plt.figure(figsize=(4, 4))
plt.xlim(-2.5, 5.5)
plt.ylim(-2.5, 5)
plt.plot(xi, yi, '*g', xi, y_gorrito, 'or', t, xbarra[0,0]*np.sin(xbarra[2,0]*t)+c, 'r')
plt.show()

# Error cometido
error = np.sum((yi - (xbarra[0, 0]*np.exp(xbarra[1,0]*xi)+xbarra[2, 0]))**2)
print('Éste es el error cuadrático:', error)
```

```python
# Método del Gradiente Descendiente (ajuste sinusoidal)

xi = np.array([0.0, 1.0, 1.5, 2.0, 2.5, 3.0, 6.0, 6.5])
yi = np.array([0.10, 0.85, 0.99, 0.91, 0.60, 0.12, -0.05, 0.42])

xbarra = np.array([[4.0], [1.0], [1]])
alpha = 0.01
gradiente = np.array([[np.sum(2*(yi-(xbarra[0,0]*np.sin(xbarra[1,0]*xi)+xbarra[2,0]))*(-np.sin(xbarra[1,0]*xi)))],
                      [np.sum(2*(yi-(xbarra[0,0]*np.sin(xbarra[1,0]*xi)+xbarra[2,0]))*(-xbarra[0,0]*np.cos(xbarra[1,0]*xi)*xi))],
                      [np.sum(2*(yi-(xbarra[0,0]*np.sin(xbarra[1,0]*xi)+xbarra[2,0]))*(-1))]])

# Tener cuidado con CÓMO SE PONEN los valores, si por columna o fila [[]]
xbarra = xbarra - alpha * gradiente/la.norm(gradiente)

# Error cometido
error = np.sum((yi - (xbarra[0, 0]*np.sin(np.exp(xbarra[1,0]*xi))+xbarra[2, 0]))**2)
# El error tiene que ser MÁS PEQUEÑO después de la primera iteración
print('Éste es el error cuadrático (INICIAL):', error)
print(xbarra, end='\n\n')

for i in range (10000):
    gradiente = np.array([[np.sum(2*(yi-(xbarra[0,0]*np.sin(xbarra[1,0]*xi)+xbarra[2,0]))*(-np.sin(xbarra[1,0]*xi)))],
                          [np.sum(2*(yi-(xbarra[0,0]*np.sin(xbarra[1,0]*xi)+xbarra[2,0]))*(-xbarra[0,0]*np.cos(xbarra[1,0]*xi)*xi))],
                          [np.sum(2*(yi-(xbarra[0,0]*np.sin(xbarra[1,0]*xi)+xbarra[2,0]))*(-1))]])

    xbarra = xbarra - alpha * gradiente/la.norm(gradiente)
    
    # Error cometido
    error = np.sum((yi - (xbarra[0, 0]*np.sin(np.exp(xbarra[1,0]*xi))+xbarra[2, 0]))**2)
    # El error tiene que ser MÁS PEQUEÑO después de la primera iteración
    # print('Éste es el error cuadrático:', error)

print()
print('Éste es el error cuadrático (FINAL):', error)
print(xbarra)
```
