## Unit 3. Vectors

- **Vector** ($v$): is a **quantity** that represents both **direction** and **magnitude** in an **n-dimensional space** (i. e., the difference of two points in *Euclidean space*). Has **tail**, **head**, **magnitude** (length) and **direction**. $P = (5, 3), \quad Q = (0, 0), \quad \vec{PQ} = (5, 3)$
- **Vector norm** ($\|v\|$): returns the magnitude of the vector.
- **Norm properties**: **no-negativity** ($\|v\|\ge 0$), **homogeneity** ($\|\alpha v\| = |\alpha| \,\|v\|$), **triangular inequality** ($\|u + v\| \le \|u\| + \|v\|$), **identity** ($\|v\| = 0 \iff v = 0$).
- **Inner product** ($\mathbf{u} \cdot \mathbf{v}$): every component in $\mathbf{u}$ multiplies the equivalent component in $\mathbf{v}$, then all are summed. The result is a **scalar number**.
- **Angle between vectors** ($\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\| \, \|\mathbf{v}\| \cos \theta$): relates the inner product to the magnitudes of the vectors and the cosine of the angle $\theta$ between them.
- **Orthogonal condition** ($\mathbf{u} \cdot \mathbf{v} = 0$): non-null vectors whose angle is $90^\circ$.
- **Normal vector**: **perpendicular** vector to a line or plane.
- **Cosine similarity** ($\operatorname{cos}(\mathbf{u}, \mathbf{v}) = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u}\| \, \|\mathbf{v}\|}$): returns a value in the interval $[-1,\,1]$ ($1$ are **parallels** in the **same direction**, $0$ are **perpendicular** and $-1$ are **parallel** in the **opposite direction**.
- **Unit vector** ($\hat v$): is a **normalized vector** of the vector **scaled to length 1**. Formally: $\hat{\mathbf{v}} = \frac{\mathbf{v}}{\|\mathbf{v}\|}$.

## Unit 4. Matrices

- **Matrix** ($A_{m \times n}$): is a **finite**, **ordered**, **rectangular array** of elements arranged in **rows** ($m$) and **columns** ($n $), where each element of the matrix is taken from a field $F$ (for example, $\mathbb{R}$ or $\mathbb{C}$). Formally: $A : \{1, 2, \ldots, m\} \times \{1, 2, \ldots, n\} \to F$, where $A(i, j) = a_{ij}$ denotes the element in the *i*-th row and *j*-th column of $A$.
- **Transpose** matrix ($A^T$): is the $n \times m$ matrix obtained by **interchanging the rows and columns** of $A$ (reflecting across its main diagonal). Formally: $A^T = [a_{ji}]$.
- Matrix **types**: **square** ($n = m$), **diagonal** (only diagonal elements), **identity** (square diagonal matrix made of zeros), **symmetric**.
- **Multiplication** ($C = AB$): it is the **dot product** of every *i-th row* and *j-th* column given their corresponding *i-th* and *j-th* locations, where very element of $C$ is calculated as $c_{ij} = \sum_{k=1}^{n} a_{ik} b_{kj}, \quad 1 \le i \le m, \; 1 \le j \le p$.
- **Addition**, **substraction** and **scalar**: typical and done element to element.
- **Determinant** of a matrix ($\|A\|$): is a **scalar value** that encodes properties of **invertibility**, **volume scaling** and **linear dependence** of rows and columns. Typical operations.
- **Adjugate** matrix ($adj(A)$): is the **transpose of the cofactor matrix of A**.  Formally: $adj(A) = C^T = [(-1)^{i+j} \det(M_{ij})]^T$, where $M_{ij}$ is the **minor** obtained by removing the *i-th* row and *j-th* column of $A$.
- **Reverse** matrix ($A^{-1}$): is the matrix such that $A A^{-1} = I$ where $I$ is the identity matrix of the same order as $A$. **Conditions** for a matrix $A$ to be invertible: must be a **square** matrix ($m = n$), have a **non-zero determinant** ($det(A) \neq 0$), and have **full rank** ($rank(A) = n$). Formally: $A^{-1} = \frac{1}{det(A)}adj(A)$.
- **Linear transformation**: is a function that mutiplies a vector by a matrix to create another vector, preserving addition and scalar multiplication. Formally: $T(\mathbf{v}) = A\mathbf{v}$, where $A$ is a matrix. Can be expressed in the form of a linear system, where $\begin{bmatrix} x' \\ y' \end{bmatrix} = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix}$ is the matricial form of the resulting equations $\begin{cases} x' = ax + by \\ y' = cx + dy \end{cases}$. As a result, a vector can be **rotated**, **scaled** (enlarged or reduced), or **reflected**, while preserving the **linear structure** of the vector space.
  - **Note**: in a **2-dimensional space**, every **movement (transformation) of the plane corresponds to a matrix transformation** and vice versa (this expression can be generalized to **n-dimensional space** and **affine transformations**).
- **Eigenvalues** and **eigenvectors** (satisfy $Av = \lambda v$): given a **non-zero vector** ($v$) that goes through a matrix ($A$) for linear transformation, **lambda** ($\lambda$) is the **scalar eigenvalue** that represents the **factor by which the eigenvector ($v$) of $\lambda$ is stretched or shrunk** under the transformation, whilst its **direction remains unchanged**. **Eigenvectors** are also called **invariant vectors**.
  - For a square matrix $A$, the **product of its eigenvalues** equals its **determinant**. Formally: $\prod_{i=1}^{n} \lambda_i = \det(A)$.
  - For a square matrix $A$, the **sum of its eigenvalues** equals its **trace**. Formally: $\operatorname{tr}(A) = \sum_{i=1}^{n} \lambda_i$.  
- **Characteristic equation** ($p(\lambda) = \det(A - \lambda I) = 0$): is the **determinantal equation used to find eigenvalues is obtained by setting the determinant to 0** . Therefore, only the values of $\lambda$ that satisfy the characteristic equation are true eigenvalues.
  - The reason the determinantal equation must be set to 0 comes from the requirement that the eigenvector $\mathbf{x} \neq 0$: $Ax = \lambda x (but, x \neq 0) \to (A-\lambda I)^{-1}(A-\lambda I)x=(A-\lambda I)^{-1}0$.
- **Characteristic polynomial** (i.e. $\lambda^2 + \lambda - 1$): is the **lambda-dependent polynomial equation** whose solutions are eigenvalues of $A$.
- Matrix **Diagonalization** (**Factorization Perspective** $M_{nxn}=PDP^{-1}$, with **linearly independent eigenvectors**): if a square matrix $M_{nxn}$ has $n$ linearly independent vectors, can be written as the product of three matrices: the **matrix of eigenvectors** ($P$), the **diagonal matrix of eigenvalues** ($D$), and the **inverse of $P$**.  
  - The **eigenvectors** come from solving the **characteristic equation** $(A - \lambda I) = 0$, and then **stacked as columns** into $P$ ($P = [v_1, v_2, \dots, v_n]$), where each $\lambda_i$ in $D$ corresponds to its matching column in $P$.  
  - Multiplying $A$ by $P$ gives $AP = PD$, meaning that $A$ acts on its eigenvectors by simply **scaling each by its eigenvalue**. Multiplying both sides by $P^{-1}$ leads to the diagonalization formula: $A = P D P^{-1}$.
  - **Use**: instead of repeatedly multiplying $A^n$, we can just **raise the diagonal matrix $D$ to a power**. Formally: $M^n=PD^nP^{-1}$.
- ***Conversion to square matrix***: a matrix $M \in \mathbb{R}^{m \times n}$ can form square matrices through $M^T M \in \mathbb{R}^{n \times n}$ and $M M^T \in \mathbb{R}^{m \times m}$ ($nxn$ and $mxm$ will have different matrix sizes). The resulting matrices share the **same nonzero eigenvalues**, which are **always positive** due to being formed by squared items.
- **Similarity matrix** ($S$): is a matrix used to **measure the correlation** between vectors by computing the **dot product of its normalized vectors**. Formally: $S = (M_{\text{norm}})^T M_{\text{norm}}, \quad \text{with } M_{\text{norm}} = \frac{M}{\|M\|}$.
- **Singular Value Decomposition (SVD, $A_{m\times n} = U \Sigma V^T$)**: is a **matrix factorization technique** used to express any given matrix in three matrices, where:  
  - $U$: is an $m\times m$ **orthogonal matrix** whose columns are the **left singular vectors** of $A$. Formally: **eigenvectors** of $MM^T$. **ROTATION**.
  - $\Sigma$: is an $m\times n$ **diagonal matrix** containing the **singular values** of $A$ (non-negative and usually sorted in descending order). Formally: **square roots of non-zero eigenvalues** of $M^TM$ or $MM^T$. **SCALING**.
  - $V^T$: is the transpose of an $n\times n$ **orthogonal matrix** whose columns are the **right singular vectors** of $A$. Formally: **eigenvectors** of $M^TM$. **ROTATION**.
  - **Interpretation**: $A$ projects vectors onto the singular vector basis ($V$), scales by $\Sigma$, and rotates via $U$.
  - **Use**: **dimensionality reduction, pseudoinverse computation, low-rank approximations, and noise filtering**..

```python
import numpy as np
import numpy.linalg as la

X = np.array([[2, -4],
              [1, 2]])
Y = np.array([[1, 1],
             [2, 5]])
print('X:\n', X)
print('Y:\n', Y)
print('X + Y:\n', X + Y)
print('X / Y:\n', X / Y)
print('XY:\n', X @ Y)
I = np.eye(2)
print('I:\n', I)

print('X^-1:\n', ln.inv(X))
print('X^-1, manually:')
detX = la.det(X)
adjX = np.array([[X[1, 1], -X[0, 1]],
                  [-X[1, 0], X[0, 0]]])
invX = (1 / detX) * adjX
print(invX)
print('AA^-1 = I:\n', X @ invX, end='\n\n')

A = np.array([[3, -2, 0], [-2, -1, 0], [0, 0, 1]])
print('A:\n', A, sep='')
eigenvalues, eigenvectors = la.eig(A)
print('Product of eigenvalues:', np.prod(eigenvalues))
print('Determinant           :', round(la.det(A), 1))
print('Sum of eigenvalues    :', round(np.sum(eigenvalues), 1))
print('Trace                 :', np.trace(A), end='\n\n')

B = np.array([[1, 0, 0, 0, 2], [0, 0, 3, 0, 0], [0, 0, 0, 0, 0], [0, 2, 0, 0, 0]])
print('B:\n', B, sep='')
print('Eigenvalues B^TB:\n', la.eig(B.T@B)[0], sep='')
print('Eigenvalues BB^T:\n', la.eig(B@B.T)[0], sep='')
U, S, V = la.svd(B, full_matrices=False)
print('U:\n', U, sep='')
print('S:\n', S, sep='')
print('V:\n', V, sep='')
print('U*S*V:\n', U@np.diag(S)@V, sep='')
```

## Unit 5. Derivatives and Gradient Descent Method

- **Limits** ($\lim_{x \to \alpha} f(x) = \alpha$): are mathematical tools used to describe the **behavior of a function as its input approaches a particular value**, especially when the function **cannot be explicitly defined** at that point, such as $f(x) = \frac{1}{x}$, where the approach to 0 can be observed but will at that point **undefined**.
- **Average rate of change** near a point ($\frac{\Delta y}{\Delta x}$): is the ratio of change in $y$ over change in $x$, which allows the observer to know the **rate of change at a certain point for a given function**.
  - Formally: $\frac{\Delta y}{\Delta x} = \frac{f(a + \Delta x) - f(a)}{\Delta x}$.
  - Geometrically: the **slope of the secant line** which passes through **two points** of the curve, which are $(a, f(a))$ and $(a + \Delta x, f(a + \Delta x))$.
- **Instantaneous rate of change** (differential coefficient): is the **limit of the average rate at which a function is changing at an exact point**, which allows the observer to know **how fast he function is changing** at one point.
  - Formally: $f'(a) = \lim_{\Delta x \to 0} \frac{f(a + \Delta x) - f(a)}{\Delta x}$.
  - Geometrically: the **slope of the tangent line to the curve** at a **single point** $(a, f(a))$, such that $f'(a) = \lim_{x \to a}\frac{f(x) - f(a)}{x - a}$.
- **Differential Method** ($dx$): uses **differential coefficients** to, either **estimate small changes** in the function or **approximate values**.
- **Basic Derivative Properties**: **Constant** Rule, **Sum/Diff** Rule, **Product** Rule, **Quotient** Rule (**change of variable**), **Chain** Rule, **Power** Rule, **Logarithmic** Rule, **General Logarithm** Rule, **Logarithmic Differentiation**.
- **Partial Derivatives** ($\frac{f(x, y)}{\partial x}$): is the rate of change of a function with respect to a certain variable whilst the others remain constant.Can be **n-order** ($\frac{\partial^2 f}{\partial x^2}$) to the same variable or **mixed partial** ($\frac{\partial^2 f}{\partial x \, \partial y}$).
- **Derivative of the Sigmoid Function**: $\sigma(x) = \frac{1}{1 + e^{-x}}$.
- **Gradient Descent Method**: is an **iterative optimization algorithm** used to find the **minimum** by **repeatedly moving in the direction opposite to the gradient** of the function *at the current point*, since the gradient points toward the direction of **maximum increase**.
  - Formally: $\mathbf{x}_{k+1} = \mathbf{x}_k - \eta \nabla f(\mathbf{x}_k)$:
    - $\mathbf{x}_k$ is the current point (parameter vector),
    - $\nabla f(\mathbf{x}_k)$ is the **gradient** of $f$ at $\mathbf{x}_k$,
    - $\eta > 0$ is the **learning rate**, which controls the step size.
  - **Goal**: to minimize the **distance between the real points and the approximated points**, usually via the **squared error**, calculated formally: $\text{Error} = \sum_i (y_i - \hat{y}_i)^2 = \sum_i (y_i - f(x_i))^2$, which works for both **linear and nonlinear models**.
    - **Visual interpretation**: pointing toward the **direction of maximum increase** of the function and moving **opposite** to it, which decreases the function most rapidly.  
  - **Update rule**: starting from a seed point, adjust parameters by subtracting **α times the normalized gradient**, formally: $\mathbf{x}_{k+1} = \mathbf{x}_k - \alpha \frac{\nabla f(\mathbf{x}_k)}{\|\nabla f(\mathbf{x}_k)\|}$. If the current step increases the error, move in the **opposite direction**.  
  - **Iterative process**: repeat until the **squared error is minimized** or the updates become sufficiently small.

## Unit 6. Model fitting and Evaluation

- **Data-based model**: is used to **predict variables** and to **understand** and **modelize relationships between variables**... In such models, **data tables** are **treated as a set of vectors**, where each vector represents an observation or instance of the variables.
- **Requirements**:
  - **Mathematical**: vector algebra, similarity between vectors, distances, etc.
  - **Basic understanding of model development**: data separation (training and validation), identifying relevant variables and statistical testing to detect overfitting and significant differences between models.
  - **Understanding of mathematical procedures** used to fit models.
- **Linear model** ($y = ax + b$): finds new variables by **linearly combining** other variables, using **sums and multiplications of constant coefficients**.
- **Several-variable linear models** ($y = a + b x_1 + c y^2 + d \log(z)$): **non-linear transformations of individual variables can be applied**, as long as the model combines the transformed variables **linearly**.
- **Data normalization**: is the process of **scaling variables to a common range or distribution**.
  - **Min-max normalization** ($\bar{x} = \frac{x - x_{\min}}{x_{\max} - x_{\min}}$): scales values to the range $[0, 1]$. Sensitive to outliers.
  - **Z-score standarization** ($x = \frac{x - x_{\text{mean}}}{x_{\text{std}}}$): **centers $x$ at 0** with a **standard deviation of 1**. Assumes approximately normal distribution.
- ***Model Fitting and Evaluation Procedure***:
  1. **Normalize the data** to scale variables to a common range or distribution.
  2. **Split the data** into training and validation sets.  
  3. **Specify the model** to be used.  
  4. **Fit the model** for an **overdetermined system of equations** ($A x = b$) using the **least squares fitting** method, formally: $x = (A^T A)^{-1} A^T b$, to find the value of $x$ that minimizes $\|A x - b\|$.  
  5. **Compute the error** by applying the **sum of squared differences**, formally: $(y - y')(y - y')^T$, to find the minimal error.
     - **Training error** $E_e^2 = (y_e - y_e')(y_e - y_e')^T$:
     - **Validation error** $E^2 = (y - y')(y - y')^T$:
  6. **Compare errors** ($\frac{E_e^2}{E^2}$): since the sum of squared differences follows a **chi-squared distribution** with $n - 1$ degrees of freedom (for $n$ data points), the result of this comparison helps to detect **overfitting**.
     - $\frac{E_e^2}{E^2} \gg 1$: $E$ is much smaller. Rare, but acceptable.  
     - $\frac{E_e^2}{E^2} \to 1$: indicates a good fit.  
     - $\frac{E_e^2}{E^2} \to 0$: $E$ is much larger, suggesting **overfitting** (the model fits **too well** to the training data and **generalizes poorly**).
  7. **visualize percentiles** (using the *Fischer distribution*): the ratio $\frac{E_e^2}{E^2}$ can be compared to the **Fischer (F-) distribution**, which shows the expected variation of error ratios for a well-generalized model.  
     - **Percentile 50**: normal. Training and validation errors are typical.  
     - **Percentile 3 (low)**: suggests **overfitting**. Training error is unusually small compared to validation one.  
     - **Percentile 96 (high)**: rare but possible. Training error is unusually high relative to validation.
  8. **Asses confidence level (%) for statistical error**: indicates the probability that the observed error falls within the expected range due to random variation.
     - **95% confidence**: the model error is expected to lie within the predicted range 95% of the time. This is the standard choice for reliable fits.
  9. **Compare models**: the ratio of errors between two models can be visualized using a **Fischer (F-) distribution**.  
     - **Percentile 50**: indicates similar performance between models.  
     - **Percentile close to 0**: model $E_1^2$ is significantly better than $E_2^2$ at a $1-p$ probability.  
     - **Percentile close to 100**: model $E_1^2$ is significantly better than $E_2^2$ at a $p$ probability.
  10. **Evaluate confidence between models** (e.g., 90% confidence):  
     - **Close to 0%**: $M_1$ is better at 90% confidence.  
     - **Close to 100%**: $M_2$ is better at 90% confidence.  
     - **Between 5% and 95%**: no significant differences between models.  
     - **Note**: if the confidence level is raised to 95%, the differences are evaluated at the 2.5% and 97.5% thresholds.

- Memory is **expensive**: Numpy is implemented in C for maximum **optimization**. Depending on the algorithm there are **memory** or **speed** costs.

- - Data and algorithms: **finding data** is important, it can be speedy but at a **memory cost**. In addition, **allocating data** increases the difficulty of the problem.  
- Library **Numpy**: implements **multidimensional array objects**, **matrixes**, ... Supports **discrete Fourier transforms**, **basic algebra, statistics, simulations...**.It is implemented in C, uses **element-by-element operations**.

So, only use copy when you need to modify the previous object and still retain the original. Do **not use copies when you are not modifying** the object.  
- **Memory usage**: if there is a string array, when a long list made of small words has a very long word, every *box* will be as big as the longest word, therefore **wasting memory space**. Also the **data type can create waste (such as saving numbers as strings)**. It takes a lot to do a `reshape` than to `create` an array.
